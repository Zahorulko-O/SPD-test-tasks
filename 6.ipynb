# №6 
#### Уявімо що в нас є задача по виявленню того які відхилення від нормованого руху допускає водій під час їзди. В нас є наступні дані: дані з осей акселерометра встановленого в мобільному, що заміряв під час руху, а також GPS координати руху автомобіля. Які ідеї як можна використати ці дані для вирішення задачі? Які ідеї щоб зробити це ефективніше (можливо поміркувати над додатковими даними)? (з кодом чи без)
 


 
Акселерометр в смартфоні – це датчик, який визначає кут нахилу електронного пристрою по відношенню до земної поверхні. величина прискорення, обчислюється наступним чином: √(x^2+y^2+z^2 )

GPS трекер - електронний пристрій для отримання інформації про місце розташування об'єкта.

Перед пошуком аномалій необхідно привести показники Додати час зняття показників і об’єднати їх в одну систему.
Пошук аномалій (відхилень від нормованого руху):
Швидкість:
1.	Порахувати різницю між точками та часом їх заміру, порахувати  швидкість руху, таким чином можна визначити середню швидкість  від неї відштовхуючись шукати прискорення та уповільнення (зупинку) 
2.	При зупинці відстань (різниця) між точками повинна зменшуватись поступово, при різних непередбачуваних ситуаціях (аварії, зіткненні, вибіганні пішохода/кота/собаки на проїжджу частину) різко.  
3.	При різкій зупинці можна визначити чи це плановене тормозіння виконане натиском на гальма чи це відбулось зіткнення з якимось предметом (при плановому повинен бути тормозний шлях з затуханням швидкості, а при зіткненні буде різка зміна швидкості на 0)
4.	При визначенні швидкості і маневреності (в населених пунктах автомобіль часто зупиняється на світлофорах, пропускає пішоходів на зебрі, повертає) можна визначити де їде машина – на трасі чи в населеному пункті, відповідно визначити чи задана швидкість є порушенням правил
5.	По манері їзди можна спробувати визначити стан адекватності водія - алкогольного сп’яніння: наскільки  різкий/агресивний, неадекватний рух машини. Для цього треба звертати увагу на різкі зміни швидкості, зміна амплітуди, троекторії руху
6.	По заносах на дорозі (їх характеру) можна спробувати визначити технічний стан автомобіля
7.	Визначивши чи лежить телефон на одному місці (змінюватись може тільки в напрямку 2 вісей) чи ні – рухається по салону (3 вісей), можна визначити стан дороги
8.	Порахувати наскільки заносить транспортний засіб (якщо рух вперед – це рух по осі у, а вправо-вліво – по осі х, то на скільки відхиляються показники по х, визначити таким чином амплітуду)
9.	Порівняти значення заносів з габаритами самої дороги(для  України беремо максимальну ширину смуги, що не повинна перевищувати 3.75м, згідно т.5.1, ДБН А.2.3-5) та машини, визначити на скільки вони різняться, чи виїжджає нашина за межі смуги
10.	Визначити повороти наліво та направо, що відбуваються за рахунок зміни напрямку руху транспортного засобу (кардинальна  зміна значень відповідно двох осей)
11.	По зміні положення по осях х та z можна відслідити перевертання машини навколо осі у
12.	При зміні положення тільки по осі  z (зніми по інших осях не виходять за нормовані) можна відслідити стрибок машини (ями)
13.	При поворотах навколо осі z може відбутись розворот на 180 або 360 градусів.
14.	Порівняти швидкість руху з середньою (чи дуже швидко, чи дуже повільно)
15.	Якщо надано дані про певний проміжок часу в днях/місяцях можна дізнатись на який колір їхала машина і чи є порушення правил дорожнього руху (або якщо буде відомо графік роботи світлофорів)
16.	Визначити скільки часу і, відповідно, скільки км проїзджає водій в день/неділю/місяць, вихідні/будні, визначити час доби та пори року.
