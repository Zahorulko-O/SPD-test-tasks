Додатково: Яка будова звука з точки зору математики?


Звук — фізичне явище, коливальний рух частинок середовища, що поширюється у вигляді хвиль у газі, рідині чи твердому тілі.
Інтенсивністю звуку (силою звуку) (I) називають величину, яку визначають середньої за часом енергією (⟨E⟩), яку переносять хвилею звуку за одиницю часу, через одиничну площадку, нормальну до напряму поширення хвилі.
Шви́дкість зву́ку — швидкість поширення акустичних (пружних) хвиль у середовищі.
Рух частинок середовища в рідинах і газах описується рівняннями Ейлера, неперервності і рівнянням стану для адіабатичного процесу.

Частота є головною математичною характеристикою звуку, що визначає його сприйняття людиною на слух.
Темп мовлення — це швидкість потоку зв'язного мовлення. 
Темп мовлення визначається передусім  швидкістю артикулювання складів і слів. Саме тому для математичного обчислення темпу мовлення найчастіше послуговуються цими одиницями, причому застосування складів, як свідчить переважна більшість фонетичних досліджень, забезпечує вищу точність вимірювання, ніж слів[1]. 
Математично темп мовлення виражається кількістю складів, вимовлених за секунду (рідше — кількістю слів за хвилину; є також практика вимірювання темпу середньою тривалістю складів чи слів). Орієнтовний діапазон темпу мовлення людини становить від 2 до 12 скл./с.

Для вирішення задачі по розрахунку швидкості (темпу) мовлення людини необхідно конвертувати аудіодоружку в текст, порахукати кількість слів/складів в цьому тексті та з аудіодорижки визначити її тривалість (в секундах). І, на кінець, поділити кількість слів/складів на тривалість ([ скл./с] / [ слів/с]).
Якщо рахувати темп по складах, то треба визначити мову, потім:
•	створити модель та навчати її на розміченому наборі даних з готовими розбиттями по складах(тільки треба врахувати, що не  в кожній мові слова можна розбити на склади і взагалі мову), порахувати кількість складів і порахувати  темп  аналогічно зі словами;
•	або прописати функцію, яка буде описувати правила поділу слів на склади (тільки треба врахувати, що не  в кожній мові слова можна розбити на склади і мову), порахувати кількість складів і порахувати  темп  аналогічно зі словами.

```python
%%time
import speech_recognition as sr 
#import pyttsx3  
import wave
import contextlib
import numpy as np
#SpeechRecognition формати: *WAV(в форматі PCM/LPCM), *AIFF, *AIFF-C,*FLAC(OGG-FLAC не підтримується)

def s_speed(audio_f):
    r = sr.Recognizer()
    harvard = sr.AudioFile(audio_f)
    print(harvard)
    with harvard as source:
        audi = r.record(source)
   # print(r.recognize_google(audi)) - озвучений текст
    n = len(r.recognize_google(audi).split()) #кількість озвучених слів
    
    with contextlib.closing(wave.open(audio_f,'r')) as f: 
        frames = f.getnframes()
        print(frames)
        rate = f.getframerate()
        t =(f.getnframes()/ float(f.getframerate()))  #час викнання аудіодоріжки
    return print("Швидкість(темп) мовлення заданого тексту складає: ", np.round(n/t), " (слів/с)")
s_speed('OSR_us_000_0011_8k.wav')    
```
