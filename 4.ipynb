```python
import re

def f_enter(T,A):
    Task = re.sub(r'[:;,\s]',' ', str(T).strip('.\!\?')).split()
    Answer = re.sub(r'[:;,\s]',' ', str(A).strip('.\!\?')).split()

    insertions = [i for i in Answer if i not in Task]

    common_for_both_texts = [i for i in Task if i in Answer]

    penalty = len(insertions)
    point = len(common_for_both_texts) - penalty 
    print(' point: ', point)
    
    score = round((3/ len(Task))* point, 2)
    print(' score: ',score) 
#f_enter(task, answer)
```
