```python
#for text processing
import re
from termcolor import colored, cprint
#cprint('Вывод с помощью cprint', 'red')
```

```python
#1  Список зі словами, що були у завданні, але в відповіді їх не було (omissions).
def f_omissions(T, A):
    Task = re.sub(r'[:;,\s]',' ', str(T).strip('.\!\?')).split()
    Answer = re.sub(r'[:;,\s]',' ', str(A).strip('.\!\?')).split()
    omissions = [i for i in Task  if i not in Answer]
    return  omissions
```

```python
#2 Список з парами слів, що були змінені місцями (swaps)
def f_swaps(T, A):
    Task = re.sub(r'[:;,\s]',' ', str(T).strip('.\!\?')).split()
    Answer = re.sub(r'[:;,\s]',' ', str(A).strip('.\!\?')).split()
    swaps = [[Task[i], Answer[i]] for i in range(len(Task)) 
        if ((i<(len(Task)-1)) and (i<(len(Answer))-1)
                                and (Task[i] == Answer[i+1]) 
                                and (Task[i+1] == Answer[i]))

             or ((i<(len(Task)-1)) and (i<(len(Answer))-1) 
                                and(Task[-i-1] == Answer[-i]) 
                                and (Task[-i] == Answer[-i-1]))]    
    return swaps
```

```python
#3 Список зі словами, що були вставлені в відповіді, але їх не було у завданні (insertions).
def f_insertions(T, A):
    Task = re.sub(r'[:;,\s]',' ', str(T).strip('.\!\?')).split()
    Answer = re.sub(r'[:;,\s]',' ', str(A).strip('.\!\?')).split()
    insertions = [i for i in Answer if i not in Task]
  
    return insertions
```

```python
#4 Список з парами-замінами (replecement)
def f_replecement(T, A):
    Task = re.sub(r'[:;,\s]',' ', str(T).strip('.\!\?')).split()
    Answer = re.sub(r'[:;,\s]',' ', str(A).strip('.\!\?')).split()
    replecement = [[Task[i], Answer[i]] for i in range(len(Task)) 
                   if ( i<len(Answer)) 
                   and (Answer[i] != Task[i])]
    return  replecement 
```

```python
#5 Частину тексту, яка є спільною для обох текстів.
def f_common_for_both_texts(T, A):
    Task = re.sub(r'[:;,\s]',' ', str(T).strip('.\!\?')).split()
    Answer = re.sub(r'[:;,\s]',' ', str(A).strip('.\!\?')).split()
    common_for_both_texts = [i for i in Task if i in Answer]

    return  common_for_both_texts
```
